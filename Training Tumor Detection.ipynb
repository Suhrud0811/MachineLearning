```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread(r"test_images/0 (236).jpg",0) # will take it as gray scale image
#plt.imshow(img)
height,width=img.shape
imgFlatterned=img.reshape(height*width,1)# basically 512*512 of 1 vector
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=5)
```

```python
model.fit(imgFlatterned)
```

```python
labels=model.labels_
labels2D=labels.reshape(height,width)
```

```python
plt.imshow(labels2D)
plt.colorbar()
```

```python
import joblib
joblib.dump(model,'KMeansTumorDetect.sav')
```
